```python
from sklearn.datasets import load_iris
x,y=load_iris(return_X_y=True)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
kmeans = kmeans.fit(x)
kmeans.labels_
```

```python
y
```

```python
import matplotlib.pyplot as plt
colors=["red","green","blue"]
```

```python
# for each element/datapoint
for index, element in enumerate(x):
  plt.scatter(element[0], element[1], color = colors[kmeans.labels_[index]])

# for each center of the clusters (centroids)
for center in kmeans.cluster_centers_:
  plt.scatter(center[0],center[1], color = "black")

#make sure to save your figure!
plt.savefig("scatter.png")
```

```python
kmeans.cluster_centers_
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 3)
pca.fit(x)
```

```python
cob = pca.components_
cob
```

```python
cob @ x[0]
```

```python
# adding our new projected datapoints into a new list
data = []
for i in range(150):
  data.append(list(cob @ x[i])) # projecting each datapoint onto the principal components
  data[-1].append(y[i])
```

```python
# shuffling the data
import random
random.seed(0)
random.shuffle(data)
data
```

```python
x = []
y = []
for line in data:
  x.append(line[:-1])
  y.append(line[-1])
y
```

```python
train_x = x[:140]
test_x = x[140:]
train_y = y[:140]
test_y = y[140:]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LogReg = LogisticRegression()
LogReg = LogReg.fit(train_x, train_y)
```

```python
LogReg.score(test_x, test_y)
```
